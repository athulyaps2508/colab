```
def build_model():              # NN Model
  model = Sequential()
  model.add(Dense(10,input_shape=(2,),activation='tanh'))
  model.add(Dense(50, activation='relu'))
  model.add(Dense(7))
  model.compile(Adam(lr= 0.001), loss='mse',metrics=['accuracy'])
  return model
```
